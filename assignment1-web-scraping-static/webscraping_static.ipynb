## **Task 1: Static Website - Extract Information About the CodeTechMentor Creator**

### **URL:**  
[https://www.codetechmentor.com/p/about-me.html](https://www.codetechmentor.com/p/about-me.html)

### **Objective:**  
Write a Python script that uses the `requests` library to fetch the HTML content of the CodeTechMentor "About" page. Use `BeautifulSoup` to parse the HTML and extract the following details about the creator:

### **Requirements:**
1. **Creator's Name**: Extract the name of the CodeTechMentor creator.
2. **Introduction Text**: Retrieve and combine all the introductory text.
3. **Social Media Links**: Find social media links.

### **Expected Output Format:**
- Print the extracted details in a structured format for easy readability.
- Ensure the script handles cases where elements might be missing.
- Provide appropriate messages if any of the required information is not found.


```python
# Start Code from Here...
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL= "https://www.codetechmentor.com/p/about-me.html"
code=requests.get(URL)
print(code)
```

```python
print(code.content)
```

```python
c=BeautifulSoup(code.content, "html.parser")
print(c.prettify())
```

```python
#variables to store requirements
creators_name=""
introduction_text=""
facebook_link=""
instagram_link=""
linkedln_link=""
github_link=""
```

```python
about_section=c.find('div',id='postBody')
print(about_section.prettify())
```

```python
introduction=about_section.find('div',class_='intro')
print(introduction.prettify())
```

```python
creators_name=introduction.find('h1').find('span',class_="highlight").text
print(creators_name)
```

```python
socials=introduction.find('div',class_="social-icons")
print(socials)
```

```python
social_links = [a['href'] for a in socials.find_all('a', href=True)]
for l in social_links:
  print(l)
```

```python
facebook_link=social_links[0]
instagram_link=social_links[1]
linkedln_link=social_links[2]
github_link=social_links[3]
```

```python
print(introduction)
```

```python
# Check if intro_section exists
if introduction:
    intro_text = introduction.get_text(separator=" ", strip=True)
    print(intro_text)
else:
    print("Intro section not found!")
```

```python
#Printing Results

print("Creators Name: ",creators_name);
print("Introduction: ",intro_text);
print("FB: ",facebook_link)
print("Insta: ",instagram_link)
print("LinkedIn: ",linkedln_link)
print("Github: ",github_link)

```

## **Task 2: Static Website Data Extraction (PriceOye Samsung Tablets)**

### **URL:**  
[https://priceoye.pk/tablets/samsung](https://priceoye.pk/tablets/samsung)

### **Objective:**  
Write a Python script that uses the `requests` library to retrieve the HTML content of the PriceOye Samsung tablets webpage. Use `BeautifulSoup` to parse the HTML and extract the following details:

### **Requirements:**
1. **Tablet Names**: Extract the names of all Samsung tablets.
2. **Prices**: Retrieve the listed prices of the tablets.
3. **Discounts**: Extract any discount information available.
4. **Exclusions**: Ensure that any products marked as **discontinued** are excluded.

### **Expected Output Format:**
- Print the extracted details in a structured format for easy readability.

```python
# Start Code from Here...
URL= "https://priceoye.pk/tablets/samsung"
code=requests.get(URL)
print(code.content)
```

```python
c=BeautifulSoup(code.content, "html.parser")
print(c.prettify())
```

```python
tablets_list=c.find('div',class_="product-list").find_all('div',class_="productBox b-productBox")
print(len(tablets_list))
```

```python
active_tablets = []
for tablet in tablets_list:
    anchor = tablet.find("a")
    if anchor and "fadeProduct" not in anchor.get("class", []):
        active_tablets.append(tablet)

print(len(active_tablets))
```

```python
tablet_names=[]
current_prices=[]
original_prices=[]
discount_percentage=[]
discount_type=[]
```

```python
for tablet in active_tablets:
  img = tablet.find("img")
  div = tablet.find("div")

  tablet_names.append(tablet.find('div',class_="p-title bold h5").text.strip())
  current_prices.append(tablet.find('div',class_="price-box p1").text.strip())

  #discount
  if img and "fixed" in img.get("class", []):
    discount_type.append("Cricket Fever Sale");
  elif div and "price-diff-retail"in div.get("class", []):
    discount_type.append("Normal Sale");
  else:
    discount_type.append("No Discount");


  if div and "price-diff-saving" not in div.get("class", []):
    discount_percentage.append("N/A");
  else:
    discount_percentage.append(tablet.find('div',class_="price-diff-saving").text.strip());


  if div and "price-diff-retail" not in div.get("class", []):
    original_prices.append("Same as Current Price");
  else:
    original_prices.append(tablet.find('div',class_="price-diff-retail").text.strip())

print(len(tablet_names))
print(len(current_prices))
print(len(original_prices))
print(len(discount_percentage))
print(len(discount_type))
```

```python
samsung_tablets_data=pd.DataFrame({'TabletName':tablet_names,
                  'CurrentPrice':current_prices,
                  'OriginalPrice':original_prices,
                  'DiscountPercentage':discount_percentage,
                  'DiscountType':discount_type,})
```
